```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.linspace(0,5,20)
b=a*a
```

```python
plt.plot(a,b)
plt.show()
plt.bar(a,b)
plt.show()
plt.scatter(a,b)
plt.show()
```

# 图片风格

```python
print(plt.style.available)
plt.style.use('ggplot')#dark_backgroud fivethirtyeight ggplot不错
plt.plot(a,b)
plt.show()
```

# 曲线风格

```python
plt.plot(a,b,'r--')
plt.show()
```

```python
# linewidth参数是线条的宽度，color是颜色，linestype是线条的类型，marker是标记的类型，makercolor是标记的颜色，
# markersize是标记大小，alpha是透明度
plt.plot(a,b,linewidth=5, color='b',linestyle='-',marker = 's',markerfacecolor='r',markersize = 6,alpha = 0.4)
```

# 子图

plt.subplot(nrows, ncols, index, **kwargs),几行，几列，第几个。只是指位置，当都小于10时，可以连着写一个数

```python
# 321 表示一会要画的图是3行2列的 最后一个1表示的是子图当中的第1个图,只是指位置
plt.subplot(3,2,1) 
plt.plot(a,b,color='r')
# 321 表示一会要画的图是3行2列的 最后一个4表示的是子图当中的第4个图
plt.subplot(324) 
plt.plot(a,b,color='b')
plt.subplot(338) 
plt.plot(a,b,color='k')
```

用subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)直接建立一系列图,sharex和sharey可以使几个图共享坐标系x和y

```python
fig,ax=plt.subplots(2,3,False,True)
ax[0,0].plot(a,b,'r-')
ax[1,2].scatter(a,b)
ax[0,1].text(0.5,0.5,'haha')
```

# 文字注释

```python
plt.plot(a,b,color='b',linestyle='-',marker = 'o',markerfacecolor='r',markersize = 3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x^2')
plt.text(0,0,'original',fontsize = 32,color='b') #（0,0）是text的位置
plt.grid(True) #这个是网格
# xy是箭头指向的位置，xytext是注释的位置，arrowprops是指定箭头的属性
plt.annotate('x=3',xy=(3,9),xytext=(1,15),arrowprops = dict(facecolor='red',shrink=0.05,headlength= 20,headwidth = 20))
```

#  3D图

```python
from mpl_toolkits.mplot3d import Axes3D
ax=plt.subplot(111,projection='3d')
ax.plot(a,b,b,'r-')
ax.scatter(a,b,b,c='b')#形参(self, xs, ys, zs=0, zdir='z', s=20, c=None, depthshade=True, \*args, \*\*kwargs)，s表示点大小，c表示颜色，两个都能用等长列表类赋值，depthshade表示深度可用透明度来表示
ax.plot(a,b,zdir='x',color='b',linestyle='-')#zdir 用来在3维图中画二维图，zdir是2维平面的法线轴
ax.bar(a,b)#plot,scatter,bar只赋值2个维度，就依然在xy平面上画
```

## 3d面

```python
#from matplotlib import cm
x=np.linspace(-5,5,11)
y=np.linspace(-5,5,11)
x,y=np.meshgrid(x,y)
print(x)
print(y)
z1=x*x+y*y
z2=z1+5
ax=plt.subplot(111,projection='3d')
ax.plot_surface(x,y,z1,rstride=2,cstride=2,cmap=plt.cm.coolwarm)#参数里rstride和cstride表示x，y参数的跨度，2就是每2个取一次值，cmap是配色体系
ax.plot_wireframe(x,y,z2,color='g',alpha=0.4)#alpha是透明度
```
